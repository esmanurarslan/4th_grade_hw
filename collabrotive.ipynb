```python
# reading the ratings dataset
ratings = pd.read_csv('/content/drive/MyDrive/Data_Mining/movielens/ratings.csv')
```

```python
# creating a pivot table
pd.pivot_table(ratings, index='userId', columns='movieId', values='rating')
```

```python
# filling the missing values in the pivot table with 0 and converting to DataFrame
df = pd.DataFrame(pd.pivot_table(ratings, index='userId', columns='movieId', values='rating', fill_value=0))
df
```

```python
# importing the packages
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

**user-based approach**

```python
# computing pairwise distances using cosine similarity between the users
distance_matrix = pairwise_distances(df, metric='cosine')
```

```python
# finding the top 5 most similar users
(distance_matrix[9]).argsort()[1:6]+1
```

```python
# pairwise_distances ile kullanıcılar arasındaki cosine distance hesaplıyoruz
distance_matrix = pairwise_distances(df, metric='cosine')

# distance_matrix numpy array dönecektir, bu yüzden pandas DataFrame'e çevirelim
distance_df = pd.DataFrame(distance_matrix, index=df.index, columns=df.index)

# Kullanıcı 10 ve kullanıcı 159 arasındaki benzerliği kontrol edelim (benzerlik = 1 - cosine distance)
similarity_score = 1 - distance_df.loc[10, 159]
print(f"Cosine similarity between User 10 and User 159: {similarity_score:.2f}")
```

```python
# finding the intersection of the sets of the movies that both users 10 and 159 have watched
set(ratings.loc[ratings.userId==10, 'movieId']).intersection(set(ratings.loc[ratings.userId==159, 'movieId']))
```

```python
# Verilen movieId'leri filtrele
movie_ids_to_check = {296, 356, 597, 1088, 1307, 1784, 2671, 2858, 2959, 4246, 4306,
                      4447, 5377, 5957, 6377, 6942, 8533, 8969, 31685, 40819, 47099,
                      49286, 56367, 64969, 68954, 72998, 81845, 84374, 88810, 95167,
                      104374, 106696, 106782, 140110}

# Sadece ilgili movieId'leri ve kullanıcıları seç
filtered_ratings = ratings[(ratings['movieId'].isin(movie_ids_to_check)) &
                           (ratings['userId'].isin([10, 159]))]

# Filmleri ve puanları birleştir
merged_data = pd.merge(filtered_ratings, movies, on="movieId")

# Çıktıyı formatlayarak grupla
for movie_id, group in merged_data.groupby("movieId"):
    title = group["title"].iloc[0]
    print(f"Movie ID {movie_id}: {title}")
    for _, row in group.iterrows():
        print(f"Rating by User {row['userId']}: {row['rating']}")
    print()  # Boşluk bırak
```

```python

# Kullanıcı 159'un rating değerinin 3.5 üzeri olan filmleri seçiyoruz
user_159_movies = ratings[ratings['userId'] == 159]
user_159_high_rating = user_159_movies[user_159_movies['rating'] > 3.5]

# Kullanıcı 10'un izlediği ve kullanıcı 159'un izlediği ortak filmleri bulalım
common_movies = pd.merge(ratings[ratings['userId'] == 10], user_159_high_rating, on="movieId")

# Kullanıcı 159'un 3.5 üzeri puan verdiği ve kullanıcı 10'un izlemediği filmleri öneri olarak alalım
unwatched_movies_by_user_10 = user_159_high_rating[~user_159_high_rating['movieId'].isin(common_movies['movieId'])]

# Önerilen filmleri movie dataframe'inde başlık bilgisiyle birleştirelim
recommended_movies = unwatched_movies_by_user_10[['movieId', 'rating']].merge(movies[['movieId', 'title']], on='movieId')

# Önerilen filmleri yazdıralım
print("Recommended Movies for User 10 based on User 159's ratings (rating > 3.5):")
print(recommended_movies.sort_values(by='rating', ascending=False))
```
